```python
 #Create category_columns and numeric_columns variables
numeric_columns = ['Population_millions', 'Working_age_population', 'Participation_rate', 'Unemployment_rate','Men_15_24', 'women_15_24', 'Men_25_64',
                   'Women_25_64', 'Average_schooling_years_total', 'Skills_as_a_constraint', 'Ed_attainment_Reading', 'Ed_attainment_Mathamatics', 
                   'Ed_attainment_Science', 'employment_sector_primary', 'employment_sector_secondary', 'employment_sector_tertiary', 
                   'employment_civil_service', 'Wage_employment', 'Self_employment', 'Farming', 'Employment_urban_areas', 'Accountant','Chemical_engineer', 
                   'Bus_driver','Sewing_machine_operator','Working_poor_below_1.25_PPP_dollars_a_day', 'Satisfied_with_life_Employed',
                   'Satisfied_with_life_Unemployed', 'Satisfied_with_life_Out_of_Labor_Force', 'Labor_share_national_income', 
                   'Gender_gap_earnings_w_m', 'Value_added_per_worker_primary', 'Value_added_per_worker_secondary', 'Value_added_per_worker_tertiary', 
                   'Civic_participation_employed', 'Civic_participation_unemployed', 'Civic_participation_out_of_labor_force', 'Wage_inEquality_90_10_ratio', 
                   'Youth_unemployment_Men', 'Youth_unemployment_woman', 'Youth_not_in_school_or_at_work_Men', 'Youth_not_in_school_or_at_work_women', 
                   'Core_ILO_conventions_ratified', 'Minimum_wage_2005_dollars_per_year', 'Separation_cost_weeks_of_salary_after_one_year', 
                   'Separation_cost_weeks_of_salary_after_10_years', 'Union_Membership_wage_employment', 'Labor_market_policy_GDP', 
                   'Social_security_contributions_salary','Social_security_coverage', 'conventions_on_migration_ratified', 'Immigrants_population', 
                   'Emmigrants_native_population', 'Remittances_GDP']
category_columns = ['Country']
```

```python
df
```

```python
# Create dummy variables for the category_columns and merge on the numeric_columns to create an X dataset
# Define target data
# Create category_columns and numeric_columns variables
X = df[numeric_columns].merge(pd.get_dummies(df[category_columns]), left_index=True, right_index=True)
X 
```

```python
# Fill in missing values in X with zeroes
X = X.fillna(0)
y = df['Core_ILO_conventions_ratified']
```

```python
pd.set_option('display.max_columns', 500)
```

```python
X.apply(lambda s: pd.to_numeric(s, errors='coerce').notnull().all())
```

```python
X.dtypes
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=60)
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
nn_model = tf.keras.models.Sequential()
```

```python
nn_model.add(tf.keras.layers.Dense(units=len(X.columns) * 2, activation = "relu", input_dim = len(X.columns)))
```

```python
nn_model.add(tf.keras.layers.Dense(units=1, activation="linear"))
```

```python
nn_model.summary()
```

```python
nn_model.compile(loss="mean_squared_error", optimizer="adam")
```

```python
nn_model.fit(X_train_scaled, np.asarray(y_train), epochs=100)
```

```python
y_train_pred = nn_model.predict(X_train_scaled)
y_test_pred = nn_model.predict(X_test_scaled)
```

```python
r2_score(y_train, y_train_pred)
```

```python
r2_score(y_test, y_test_pred)
```

```python
deep_model = tf.keras.models.Sequential()
```

```python
deep_model.add(tf.keras.layers.Dense(units=len(X.columns) *2, activation = 'relu', input_dim = len(X.columns)))
```

```python
deep_model.add(tf.keras.layers.Dense(units=len(X.columns) *2, activation = 'relu'))
```

```python
deep_model.add(tf.keras.layers.Dense(units=1, activation="linear"))
```

```python
deep_model.add(tf.keras.layers.Dense(units=1, activation="linear"))
```

```python
deep_model.summary()
```

```python
deep_model.compile(loss="mean_squared_error", optimizer="adam")
```

```python
deep_model.fit(X_train_scaled, np.asarray(y_train), epochs=100)
```

```python
y_train_pred = deep_model.predict(X_train_scaled)
y_test_pred = deep_model.predict(X_test_scaled)
```

```python
r2_score(y_train, y_train_pred)
```

```python
r2_score(y_test, y_test_pred)
```
